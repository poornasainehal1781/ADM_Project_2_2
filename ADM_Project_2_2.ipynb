```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from UCI URL
url = "https://archive.ics.uci.edu/static/public/53/data.csv"
df = pd.read_csv(url)

# Show basic structure of the dataset
print(df.head())  # Optional: Just to confirm data is loaded correctly

# Set seaborn theme
sns.set(style="whitegrid")

# 1. Pairplot - relationships between all features, colored by species
sns.pairplot(df, hue="class", height=2.5)
plt.suptitle("üîó Pairwise Feature Relationships", y=1.02)
plt.show()

# 2. Boxplot - distribution of each feature by class
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x="class", y=column, palette="Set2")
    plt.title(f"üì¶ Boxplot of {column}")
plt.tight_layout()
plt.show()

# 3. Correlation heatmap
plt.figure(figsize=(8, 6))
corr = df.iloc[:, :-1].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("üîó Feature Correlation Heatmap")
plt.show()

# 4. Countplot - number of samples per class
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="class", palette="viridis")
plt.title("üî¢ Count of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```
